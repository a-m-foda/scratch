```python
import numpy as np 
import matplotlib.pyplot as plt

circle=np.zeros((500,500), dtype='uint8')
for i in range(len(circle)):
    for j in range(len(circle)):
        if ((i-150)**2)+((j-150)**2) < 10000 :
            circle[i,j]=1
for i in range(len(circle)):
    for j in range(len(circle)):
        if i< 400 and i>300: 
            if j<400 and j>300:
                circle[i,j]=1
plt.imshow(circle,origin='lower',cmap='gray')
plt.colorbar()
plt.show()
    

                
boxy =np.zeros((500,500),dtype='uint8')# (rows,columns) or (height,width)
for i in range(len(boxy)):
    boxy[i]=np.sum(circle[i])/np.max(circle)
plt.imshow(boxy,origin='lower',cmap='gray')
plt.colorbar()
plt.show()
    
#plt.axis([min(Hproj), max(Hproj), max(arrange), 0])
#plt.show()

#create a horizontal BACK-projection directly without finding the histogram...

boxx =np.zeros((500,500),dtype='uint8')# (rows,columns) or (height,width)
for i in range(len(boxx)):
    boxx[i]=np.sum(circle,axis=0)/np.max(circle)
    
plt.imshow(boxx,origin='lower',cmap='gray')
plt.colorbar()
plt.show()

final= boxx+boxy

plt.imshow(final,origin='lower',cmap='gray')
plt.colorbar()
plt.show()
```
